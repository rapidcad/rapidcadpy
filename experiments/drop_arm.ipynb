```python
import rapidcadpy as rc
```

```python
app = rc.InventorApp()
```

```python
arm_thick = 5
width = 9

app.new_document()
wp = app.work_plane("XY")

# arm: loft rectangle at hub to rectangle at dropped tip
arm = (
    wp.move_to(-width/2, 0)
    .line_to(-width/2, 20)
    .line_to(-8, 45)
    .three_point_arc((0, 53), (8, 45))
    .line_to(width/2, 20)
    .line_to(width/2, 0)
    .three_point_arc((0, -4.5), (-width/2, 0))
    .extrude(arm_thick)
)

hole = wp.move_to(0, 45).circle(4).extrude(arm_thick)

arm.cut(hole)
```
